```python
#Imports
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
import warnings
warnings.filterwarnings("ignore")
```

```python
# Load the dataset
file_path = 'ABT_GroupI.xlsx'
data = pd.read_excel(file_path)

# Display the first few rows of the dataset
data.head()
```

```python
# print the shape of the dataset
print(data.shape)
```

```python
# check for duplicates
data.duplicated().sum()
```

- 1248 unique clients
- 40 features

```python
# display the columns of the dataset
print(data.columns)
```

```python
# delete cust_id and use 1, 2, 3 index instead
data = data.drop('cust_id', axis=1)
```

```python
# change the column name 'Type_of_costumer' to 'Type_of_customer'
data = data.rename(columns={'Type_of_costumer':'Type_of_customer'})

# change the column name 'Mon_Home and kitchen' to 'Mon_Home_and_kitchen'
data = data.rename(columns={'Mon_Home and kitchen':'Mon_Home_and_kitchen'})

# change the column name 'Spendings_Credit Card' to 'Spendings_Credit_Card'
data = data.rename(columns={'Spendings_Credit Card':'Spendings_Credit_Card'})
```

    - 'Percentage_Cash', 'Percentage_Paypal', 'Percentage_CreditCard'

```python
# Calculate mean, median, and mode for numerical columns
mean_values = data.mean(numeric_only=True)
median_values = data.median(numeric_only=True)
mode_values = data.mode(numeric_only=True).iloc[0]

# Display the results
print("Mean:\n", mean_values)
print("\nMedian:\n", median_values)
print("\nMode:\n", mode_values)
```

```python
# Check for null values and count them for each column
null_counts = data.isnull().sum()

# Display the null counts
print(null_counts)
```

```python
# Find the 10 highest spending customers
top_spenders = data.sort_values(by='Amount_of_money', ascending=False).head(10)

# Display the result
print(top_spenders)
```

```python
average_age_top_spenders = top_spenders['Age'].mean()
print("Average Age of Top Spenders:", average_age_top_spenders)
```

### Number of customers per `Type_of_customer`

```python
# check if any clients do not have one of the following types of customers
unique_types = data['Type_of_customer'].unique()
print(unique_types)

# number of people in each type of customer
type_of_customer_counts = data['Type_of_customer'].value_counts()
print(type_of_customer_counts)
```

### Total spent, average spending, city and kids per `Type_of_customer`

```python
# Define custom colors for each customer type and average
custom_palette = {"gold": "#FFD700", "silv": "#C0C0C0", "bron": "#8B4513", "average": "#4682B4", 0: "#e74c3c", 1: "#3498db"}

# Calculate overall dataset averages
overall_avg_spent = data['Amount_of_money'].mean()
overall_avg_tenure = data['Customer_Since'].mean()

# Bar plot for number of customers per customer type
plt.figure(figsize=(10, 6))
sns.barplot(x=type_of_customer_counts.index, y=type_of_customer_counts.values, palette=custom_palette, edgecolor="black")
plt.title("Number of Customers by Customer Type", fontsize=14, fontweight='bold')
plt.xlabel("Customer Type", fontsize=12)
plt.ylabel("Number of Customers", fontsize=12)
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

# Bar plot for average spending per customer type
plt.figure(figsize=(10, 6))
sns.barplot(x=data['Type_of_customer'], y=data['Amount_of_money'], palette=custom_palette, edgecolor="black")
plt.axhline(y=overall_avg_spent, color=custom_palette["average"], linestyle="--", linewidth=2, label="Average Customer")
plt.title("Average Spending per Customer Type", fontsize=14, fontweight='bold')
plt.xlabel("Customer Type", fontsize=12)
plt.ylabel("Average Amount of Money Spent", fontsize=12)
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

# Boxplot for amount of money spent across customer types
plt.figure(figsize=(10, 6))
sns.boxplot(x=data['Type_of_customer'], y=data['Amount_of_money'], palette=custom_palette)
plt.axhline(y=overall_avg_spent, color=custom_palette["average"], linestyle="--", linewidth=2, label="Average Customer")
plt.title("Distribution of Amount of Money Spent by Customer Type", fontsize=14, fontweight='bold')
plt.xlabel("Customer Type", fontsize=12)
plt.ylabel("Amount of Money Spent", fontsize=12)
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()


# Graph for the number of each customer type per City
plt.figure(figsize=(10, 6))
sns.countplot(x='City', hue='Type_of_customer', data=data, palette=custom_palette)
plt.title("Number of Customers by City and Customer Type", fontsize=14, fontweight='bold')
plt.xlabel("City", fontsize=12)
plt.ylabel("Number of Customers", fontsize=12)
plt.legend(title="Customer Type")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

# Grapth for the number of clients with and without kids per type of customer
plt.figure(figsize=(10, 6))
sns.countplot(x='Type_of_customer', hue='Kids', data=data, palette=custom_palette)
plt.title("Number of Customers by Customer Type and Kids Status", fontsize=14, fontweight='bold')
plt.xlabel("Customer Type", fontsize=12)
plt.ylabel("Number of Customers", fontsize=12)
plt.legend(title="Kids Status")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

### Average % of total spending in different departments per `Type_of_customer`

```python
# Define custom colors for each customer type and average
custom_palette = {"gold": "#FFD700", "silv": "#C0C0C0", "bron": "#8B4513", "average": "#4682B4"}

# List of features to visualize
features = [
    'Percentage_Bags', 'Percentage_Books', 'Percentage_Clothing', 'Percentage_Electronics',
    'Percentage_Footwear', 'Percentage_Home_and_kitchen'
]

# Group data by 'Type_of_customer' to create `customer_summary`
customer_summary = data.groupby('Type_of_customer').agg({feature: 'mean' for feature in features})

# Loop through each feature and create visualizations
for feature in features:
    overall_avg = data[feature].mean()  # Calculate dataset-wide mean for the feature
    
    # Bar plot for percentage distribution per customer type
    plt.figure(figsize=(10, 6))
    sns.barplot(x=customer_summary.index, y=customer_summary[feature], palette=custom_palette, edgecolor="black")
    plt.axhline(y=overall_avg, color=custom_palette["average"], linestyle="--", linewidth=2, label="Dataset Average")
    plt.title(f"Average % of total spent on {feature.replace('_', ' ')} by Customer Type", fontsize=14, fontweight='bold')
    plt.xlabel("Customer Type", fontsize=12)
    plt.ylabel(f"Average {feature.replace('_', ' ')}", fontsize=12)
    plt.legend()
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    plt.show()

```

### % spent across different departments per `Type_of_customer`

```python
# List of department percentage features
departments = [
    'Percentage_Bags', 'Percentage_Books', 'Percentage_Clothing', 'Percentage_Electronics',
    'Percentage_Footwear', 'Percentage_Home_and_kitchen'
]

# Create pie charts for each customer type
for customer_type in ["gold", "silv", "bron"]:
    plt.figure(figsize=(8, 8))
    
    # Get spending distribution for this customer type
    department_distribution = customer_summary.loc[customer_type, departments]
    
    # Create pie chart
    department_distribution.plot.pie(autopct='%1.1f%%', startangle=90, 
                                     colors=sns.color_palette("pastel", len(departments)))

    plt.title(f"Spending Distribution Across Departments for {customer_type.title()} Customers", fontsize=14, fontweight='bold')
    plt.ylabel("")  # Hide default ylabel
    plt.show()
```

### Average % spent online, in store or through a catalog per `Type_of_customer`

```python
# List of shopping channel percentage features
shopping_channels = ['Percentage_Online', 'Percentage_Store', 'Percentage_Catalog']

# Group data by 'Type_of_customer' to create `customer_summary` if not defined
customer_summary = data.groupby('Type_of_customer').agg({feature: 'mean' for feature in shopping_channels})

# Loop through each feature and create bar plots
for feature in shopping_channels:
    overall_avg = data[feature].mean()  # Calculate dataset-wide mean for the feature
    
    # Bar plot for percentage distribution per customer type
    plt.figure(figsize=(10, 6))
    sns.barplot(x=customer_summary.index, y=customer_summary[feature], palette=custom_palette, edgecolor="black")
    plt.axhline(y=overall_avg, color=custom_palette["average"], linestyle="--", linewidth=2, label="Dataset Average")
    plt.title(f"Average % spent {feature.replace('Percentage_', '')} per Customer Type", fontsize=14, fontweight='bold')
    plt.xlabel("Customer Type", fontsize=12)
    plt.ylabel(f"Average %  spent{feature.replace('Percentage_', '')}", fontsize=12)
    plt.legend()
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    plt.show()
```

```python
# List of payment method percentage features
payment_methods = ['Percentage_Cash', 'Percentage_Paypal', 'Percentage_CreditCard']

# Group data by 'Type_of_customer' to create `customer_summary` if not defined
customer_summary = data.groupby('Type_of_customer').agg({feature: 'mean' for feature in payment_methods})

# Loop through each feature and create bar plots
for feature in payment_methods:
    overall_avg = data[feature].mean()  # Calculate dataset-wide mean for the feature
    
    # Bar plot for percentage distribution per customer type
    plt.figure(figsize=(10, 6))
    sns.barplot(x=customer_summary.index, y=customer_summary[feature], palette=custom_palette, edgecolor="black")
    plt.axhline(y=overall_avg, color=custom_palette["average"], linestyle="--", linewidth=2, label="Dataset Average")
    plt.title(f"Average proportion of payments by {feature.replace('Percentage_', '')} per Customer Type", fontsize=14, fontweight='bold')
    plt.xlabel("Customer Type", fontsize=12)
    plt.ylabel(f"Average proportion of payments by {feature.replace('Percentage_', '')} per Customer Type", fontsize=12)
    plt.legend()
    plt.grid(axis="y", linestyle="--", alpha=0.7)
    plt.show()
```
